# EDA 

```python
# making sure no null values
print(df_lectures.isna().sum())
print(df_questions.isna().sum())
print(df_train.isna().sum())
print(df_test.isna().sum())
```

```python
# null values in train/test are fine for now, removing the null tag from questions
df_questions.fillna('-1', inplace=True)
df_questions.isna().sum()
```

```python
# looking at lecture types
type_ = df_lectures.copy()
type_['count'] = 1
type_ = type_.groupby('type_of')[['count']].sum().reset_index()
plt.bar(type_['type_of'], type_['count'])
plt.title('Most Common Lecture Types')
plt.show()
```

So the 'part' in lectures/questions corresponds to a higher level section of the test (ie. listening, comprehension, talks, etc), that I currently am not planning to use in the analysis.

The important column is 'tag', which corresponds to the skill(s) being tested on. These are vital to my analysis as I will be modeling a Bayesian Knowledge Tracing network per skill/tag item to track mastery of that skill for each student

```python
# displaying lecture tags
lectags = df_lectures.groupby('tag').count().drop(['part', 'type_of'], axis=1).rename(columns={'lecture_id':'count'})
lectags
```

```python
lectags.index = lectags.index.astype(int)
```

```python
# getting tag distributions for questions
tags = dict()
for i in df_questions['tags']:
    x = i.split()
    for j in x:
        j = int(j)
        tags[j] = tags.get(j, 0) + 1
```

```python
plt.figure(figsize=(20,20))
plt.barh(list(sorted(tags.keys())), tags.values(), label='Questions')
plt.barh(lectags.index, lectags['count'], label='Lectures')
plt.ylabel('tag')
plt.xlabel('count')
plt.title('Distribution of Tags')
plt.legend()
plt.show()
```

So looks like some skills have a lot of questions while others don't and lectures are pretty evenly dispersed

Now looking at questions in bundles... it looks like most questions are in their own bundle but some bundles do have multiple questions

```python
# showing bundles that have more than 1 question
dfq = df_questions.groupby('bundle_id')[['question_id']].count().reset_index().rename(columns={'question_id':'n_questions'})
print('Number of bundles with exactly 1 question:', len(dfq[dfq['n_questions'] == 1]))
print('Number of bundles with >1 question:', len(dfq[dfq['n_questions'] > 1]))
print('\nDistribution of bundle questions:')
dfq.groupby('n_questions')[['bundle_id']].count().rename(columns={'bundle_id':'count'})
```

## Now looking at training data and questions

My initial BKT model won't take into account lectures (to form a simple baseline model), and testing data won't be looked at too much prior to model evaluation

```python
# average number of questions attempted by users
# content_type_id is 0 for questions and 1 for lectures
avgq = df_train[df_train['content_type_id'] == 0]
avgq['count'] = 1
avgq = avgq.groupby('user_id')[['count']].sum()
```

```python
avgq.rename(columns={'count':'qs_attempted'}, inplace=True)
avgq['qs_attempted'].describe()
```

```python
sns.violinplot(avgq['qs_attempted'])
plt.show()
```

Upper quartile is pretty stretched out so I'm going to look at the distributions in the first 3/4 

```python
favgq = avgq[avgq['qs_attempted'] <= 154]
sns.violinplot(favgq['qs_attempted'])
plt.show()
```

```python
sns.distplot(favgq['qs_attempted'])
plt.show()
```

(kind of?) normally distributed here, though heavily skewed right. Seems to be a huge peak at ~30 questions signaling that's when a lot of people stop their learning, maybe there's some incentive in the learning module to get to 30 questions?

```python
idx = np.where((avgq['qs_attempted'] >= 30) & (avgq['qs_attempted'] <= 40))
q3040 = avgq.iloc[idx]
print('{}% of people stop their learning between 30 and 40 questions'.format(round(100 * len(q3040) / len(avgq), 3)))
```

### Looking at relationship between train and test sets now
I need to know whether the students in the test set are all in the training set

```python
# are all test students in training set?
x = []
test_stu = df_test['user_id'].unique()
train_stu = df_train['user_id'].unique()
for student in test_stu:
    if student not in train_stu:
        x.append(student)
        print(student)
print('{} student out of {} are not in the training set'.format(len(x), len(test_stu)))
```

```python
# what kind of content is in test set?
df_test['content_type_id'].unique()
```

Only questions in test set, which makes sense since the goal is to predict whether students will answer those questions are correct or not
